```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Carregar o arquivo train.csv
# Como o arquivo está dentro da pasta 'src', indicamos o caminho relativo:
df = pd.read_csv("src/train.csv")

# 2. Quantidade de linhas e colunas
print(f"O dataset possui {df.shape[0]} linhas e {df.shape[1]} colunas.")

```

```python
# 3. Tipo de dado de cada variável
print("\nTipos de dados:")
print(df.dtypes)

```

```python
# 4. Tabela de frequência e gráfico para a variável 'Survived'
# Nota: No dataset do Titanic, a coluna geralmente se chama 'Survived'
print("\nTabela de Frequência - Sobrevivência:")
freq_survival = df['Survived'].value_counts()
print(freq_survival)

# Gráfico de barras (ideal para categorias discretas)
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=df, palette='viridis')
plt.title('Distribuição de Sobreviventes (0 = Não, 1 = Sim)')
plt.show()

```

```python
# 5. Quantas pessoas sobreviveram e classe mais frequente
sobreviventes = df['Survived'].sum()
classe_frequente = df['Pclass'].mode()[0]
print(f"\nTotal de sobreviventes: {sobreviventes}")
print(f"Classe mais frequente: {classe_frequente}ª Classe")

```

```python
# 6. Relação entre 'Sex' e 'Survived' (Tabela e Gráfico)
print("\nRelação Sexo vs Sobrevivência:")
tabela_sex_surv = pd.crosstab(df['Sex'], df['Survived'])
print(tabela_sex_surv)

# Gráfico de barras agrupadas
plt.figure(figsize=(6, 4))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Sobrevivência por Sexo')
plt.show()

```

```python
# 7. Distribuição das idades
# Gráfico de Histograma com KDE (densidade) para ver a distribuição
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'].dropna(), kde=True, color='skyblue')
plt.title('Distribuição de Idade dos Passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```
