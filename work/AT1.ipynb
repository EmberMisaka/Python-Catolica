# Exercício 1 - Cadastro e análise simples de veículos

Uma concessionária deseja registrar alguns veículos disponíveis para venda.
Crie um programa em Python que crie uma lista de veículos.

Cada veículo deve ser representado por um dicionário contendo:
- marca
- modelo
- preço
- ano

Exemplo de estrutura esperada:

```
veiculo = {
    "marca": "Toyota",
    "modelo": "Corolla",
    "preco": 120000,
    "ano": 2022
}
```

O programa deve:

- Adicionar pelo *menos 5 veículos* à lista.
- Exibir todos os veículos cadastrados.
- Calcular o *preço médio dos veículos*.
- Mostrar quais *veículos custam acima da média*.

### Saída esperada (exemplo)
```
Preço médio dos veículos: 85000
Veículos acima da média:
Honda Civic - 110000
Toyota Corolla - 120000
```


````python
def sistema_carros():
    frota = []
    
    print("--- Cadastro de Veículos ---")
    
    # Loop para garantir o cadastro de pelo menos 5 veículos
    for i in range(1, 6):
        print(f"\nCadastro do {i}º veículo:")
        marca = input("Marca: ")
        modelo = input("Modelo: ")
        preco = float(input("Preço: R$ "))
        ano = int(input("Ano: "))
        
        # Criando o dicionário do carro e adicionando à lista
        carro = {
            "marca": marca,
            "modelo": modelo,
            "preco": preco,
            "ano": ano
        }
        frota.append(carro)

    # 1. Exibir todos os veículos
    print("\n" + "="*30)
    print("VEÍCULOS CADASTRADOS")
    print("="*30)
    for c in frota:
        print(f"{c['marca']} {c['modelo']} - Ano: {c['ano']} - R$ {c['preco']:.2f}")

    # 2. Calcular o preço médio
    total_precos = sum(c['preco'] for c in frota)
    media_preco = total_precos / len(frota)
    
    print("-" * 30)
    print(f"Preço médio da frota: R$ {media_preco:.2f}")
    print("-" * 30)

    # 3. Mostrar veículos acima da média
    print("\nVeículos com preço acima da média:")
    encontrou_acima = False
    for c in frota:
        if c['preco'] > media_preco:
            print(f"-> {c['modelo']} ({c['marca']}): R$ {c['preco']:.2f}")
            encontrou_acima = True
            
    if not encontrou_acima:
        print("Nenhum veículo está acima da média.")

# Executar o programa
if __name__ == "__main__":
    sistema_carros()
````

# Exercício 2 - Sistema simples de financiamento

Uma concessionária oferece financiamento de veículos.

Crie uma função chamada:

```
calcular_financiamento(preco, entrada, parcelas)
```

A função deve calcular o valor financiado.

```
valor_financiado = preco - entrada
```

Considerar uma taxa de juros simples de 1.5% ao mês.

Fórmula simplificada:
```
valor_total = valor_financiado * (1 + taxa * parcelas)
parcela = valor_total / parcelas
```
A função deve retornar o valor da parcela.


### Crie um programa que:

- Defina o preço de um veículo
- Pergunte ao usuário: valor de entrada e número de parcelas
- Use a função para calcular o financiamento
- Exiba o valor da parcela
- Impedir parcelas maiores que 72 meses
- Mostrar uma mensagem caso isso ocorra (Número máximo de parcelas é 72.)

### Exemplo de saída

```
Preço do veículo: 80000
Entrada: 20000
Parcelas: 48
Valor da parcela: 1785.50
```


````python
def calcular_financiamento(preco, entrada, parcelas):
    """
    Calcula o valor da parcela mensal com base em juros simples de 1.5%.
    """
    taxa = 0.015
    valor_financiado = preco - entrada
    
    valor_total = valor_financiado * (1 + taxa * parcelas)
    
    valor_parcela = valor_total / parcelas
    return valor_parcela

# MAIN

preco_veiculo = 80000.00
print(f"Preço do veículo: R$ {preco_veiculo:.2f}")

try:
    entrada = float(input("Digite o valor de entrada: R$ "))
    num_parcelas = int(input("Digite o número de parcelas: "))

    if num_parcelas > 72:
        print("Erro: Número máximo de parcelas é 72.")
    elif num_parcelas <= 0:
        print("Erro: O número de parcelas deve ser maior que zero.")
    elif entrada >= preco_veiculo:
        print("A entrada é maior ou igual ao preço. Não há necessidade de financiamento!")
    else:
        # 4. Calcular e exibir o resultado
        resultado_parcela = calcular_financiamento(preco_veiculo, entrada, num_parcelas)
        print(f"\n--- Resultado do Financiamento ---")
        print(f"Valor da parcela: R$ {resultado_parcela:.2f}")

except ValueError:
    print("Por favor, insira apenas números válidos.")
````
