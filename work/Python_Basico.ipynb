### Declaração de Variáveis

Python tem uma tipagem dinâmica, ou seja, ele não precisa de uma "declaração" do tipo de variável como ```int x = 1```

````python
nome = "Ember"
idade = 21
salario = 2.90
Is_Rachador = True

print(f"Nome: {nome}\n idade: {idade}\n Salario: {salario}\n VOCÊ É UM RACHADOR? {Is_Rachador}")
````

### Função (type)

Para mostrar o tipo de uma variável

````python
print(type(nome))
print(type(idade))
print(type(salario))
print(type(Is_Rachador))
````

### Operadores

Trazendo um pouco sobre operadores aritméticos e lógicos

Aritméticos > ```+```,```-```,```*```,```/```,```**```,```%```.

Lógicos > ```>```,```<```,```>=```,```<=```,```!=```,```==```.

> Aritméticos:

````python
A = 2
B = 3

soma = A + B
subtracao = A - B
divisao = A / B
potenciacao = A**B
modulo = A % B

print(soma)
print(subtracao)
print(divisao)
print(potenciacao)
print(modulo)
````

> Operadores Lógicos:

````python
print(A > B)
print(A < B)
print(A == B)
print(A != B)
````

## Estruturas Condicionais
### If/else

````python
idade = int(input("Digite sua idade: "))
possui_habilitacao = True

if (idade >= 18 & possui_habilitacao):
    print("Você é maior de idade e possui habilitacao")
elif idade >= 18:
    print("Você é maior de idade")
else:
    print("Você não é maior de idade")
````

## Estruturas de Repetição
### For


````python
for i in range (5):
    print(i)
````

  ### While

````python
cont = 0
while cont <= 10:
    print(cont)
    cont += 1
````
